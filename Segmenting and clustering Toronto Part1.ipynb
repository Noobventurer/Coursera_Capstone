```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
# Get the title to make sure we are on the right page
title = soup.title
print(title)
```

```python
#use the find_all() method of soup to extract useful html tags within a webpage, in this case I wanted to check all the table elements on the page and print them into json
table = soup.find_all('table')[0] 
df = pd.read_html(str(table))
print(df[0].to_json(orient='records'))
```

```python
#From all the json data, we can see that the info we need is present and ordered from index 0
#I decided to put the info into a panda with the index 0 as header as it was what I needed Postcode	Borough	Neighbourhood
df = pd.read_html(str(table))[0]
Postalcode = df["Postcode"].tolist()
Borough = df["Borough"].tolist()
Neighbourhood = df["Neighbourhood"].tolist()
df
```

```python
#If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough
#if the borough was not assigned, the for and if function assigned the non assigned to the name of the borough
df = df[df.Borough != 'Not assigned']
for index, row in df.iterrows():
    if row['Neighbourhood'] == 'Not assigned':
        row['Neighbourhood'] = row['Borough']

df
```

```python
# More than one neighborhood can exist in one postal code area. For example, in the table on the Wikipedia page, you will notice that M5A is listed twice and has two neighborhoods: Harbourfront and Regent Park. These two rows will be combined into one row with the neighborhoods separated with a comma as shown in row 11 in the above table.
#
df = df.groupby(['Postcode','Borough'])['Neighbourhood'].apply(lambda x: ','.join(x)).reset_index()
df
```
